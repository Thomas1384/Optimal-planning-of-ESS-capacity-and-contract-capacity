```python
from datetime import *
import pandas as pd
import numpy as np
import random
import time
import csv
import os
from sklearn.cluster import KMeans

os.chdir(r'C:\Users\li303\OneDrive - 國立成功大學 National Cheng Kung University\桌面\高港\廠商提供')

load_21 = pd.read_excel('7櫃LOAD.xlsx', sheet_name='2021')
load_21['date'] = pd.to_datetime(load_21['date'])
load_21.set_index(load_21['date'], inplace=True, drop=True)
load_21 = load_21.drop(['date'], axis=1)

load_21_scaleup = load_21*(15/7)

pv_2122 = pd.read_csv('7櫃PV_2021-22.csv')
pv_2122['date'] = pd.to_datetime(pv_2122['date'])
pv_2122.set_index(pv_2122['date'], drop=True, inplace=True)
pv_2122 = pv_2122.drop(['date'], axis=1)

pv_21 = pv_2122['2021']

pv_21.rename(columns=dict(zip(pv_21.columns, load_21_scaleup.columns)), inplace=True)
netload_21_scaleup = load_21_scaleup - pv_21

KM=KMeans(n_clusters=5,init='random',random_state=5)
KM.fit(netload_21_scaleup)
cluster_index = KM.predict(netload_21_scaleup)
cluster_index

netload_21_scaleup.insert(0, column='cluster_index', value=cluster_index)

cluster_centers = KM.cluster_centers_
netload_21_scaleup_ClusterCenters = pd.DataFrame(cluster_centers)
cluster_counts =np.bincount(cluster_index)

netload_21_scaleup_ClusterCenters.insert(0, column='days', value=cluster_counts)
netload_21_scaleup_ClusterCenters.to_csv('netload_21_scaleup_ClusterCenters.csv', encoding='utf-8_sig', index=False)
netload_21_scaleup_ClusterCenters
```

```python
from datetime import *
import pandas as pd
import numpy as np
import random
import time
import csv
import os
from sklearn.cluster import KMeans

os.chdir(r'C:\Users\li303\OneDrive - 國立成功大學 National Cheng Kung University\桌面\高港\網頁匯入檔案')

load_data = pd.read_csv('load_data.csv')
load_data = load_data.drop(['date'], axis=1)
pv_data = pd.read_csv('pv_data.csv')
pv_data = pv_data.drop(['date'], axis=1)

pv_data.rename(columns=dict(zip(pv_data.columns, load_data.columns)), inplace=True)
netload_data = load_data - pv_data

KM=KMeans(n_clusters=5,init='random',random_state=5)
KM.fit(netload_data)
cluster_index = KM.predict(netload_data)

netload_data.insert(0, column='cluster_index', value=cluster_index)

cluster_centers = KM.cluster_centers_
netload_21_scaleup_ClusterCenters = pd.DataFrame(cluster_centers)
cluster_counts =np.bincount(cluster_index)

netload_21_scaleup_ClusterCenters.insert(0, column='days', value=cluster_counts)
netload_21_scaleup_ClusterCenters.to_csv('netload_21_scaleup_ClusterCenters.csv', encoding='utf-8_sig', index=False)
netload_21_scaleup_ClusterCenters
```
